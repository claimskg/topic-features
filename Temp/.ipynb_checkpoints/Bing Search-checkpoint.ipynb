### Initialization

```python
subscription_key = "xxx"
assert subscription_key

search_url = "https://api.cognitive.microsoft.com/bing/v7.0/search"

search_term = "fact checking"
```

### Query

```python
import requests

headers = {"Ocp-Apim-Subscription-Key": subscription_key}
params = {"q": search_term, "mkt":"en-US", "count":"50", "responseFilter":"Webpages"}
response = requests.get(search_url, headers=headers, params=params)
response.raise_for_status()
search_results = response.json()
```

### Results

```python
from IPython.display import HTML

rows = "\n".join(["""<tr>
                       <td><a href=\"{0}\">{1}</a></td>
                       <td>{2}</td>
                     </tr>""".format(v["url"], v["name"], v["snippet"])
                  for v in search_results["webPages"]["value"]])
HTML("<table>{0}</table>".format(rows))
```

```python
import joblib

class Claim:
    def __init__(self, claimID):
        self.claimID = claimID
        self.claimText = ""
        self.quotations = []
        self.sources = []
        self.cues = []

test = joblib.load("Data/c1_joblib")
```

```python
for claim in test:
    print("ID: %s" % claim)
    print("Text: %s" % test[claim].claimText)
    print("Quotes: %s" % test[claim].quotations)
    print("Sources: %s" % test[claim].sources)
    print("Cues: %s" % test[claim].cues)
    print()       
```

```python
for claim in test.keys():
    if test.get(claim).quotations:
        print(claim)
        print(str(test.get(claim).__dict__))
```
